```python
#Aula 1 - código
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")

linhas, colunas = df.shape[0], df.shape[1]
print("Esse é o número de linhas: ", linhas)
print("Esse é o número de colunas: ", colunas)

renomear_colunas = {
    'work_year': 'ano_trabalho',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}
df.rename(columns = renomear_colunas, inplace=True)
df.columns

df['senioridade'] = df['senioridade'].replace({
    'SE': 'Sênior',
    'Intermediário': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
})
df['senioridade'].value_counts()

df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'Contrato': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
})
df['contrato'].value_counts()

df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})
df['tamanho_empresa'].value_counts()

df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    100: 'Totalmente Remoto',
    50: 'Híbrido'
})
df['remoto'].value_counts()

```

```python
import numpy as np
df_salarios = pd.DataFrame({
    'nome':["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    'salario':[4000, np.nan, 5000, np.nan, 100000]
})
#Preenchendo os valores nulos com a media - .mean() - com arredondamento com duas casas decimais - round(2)
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
#Preenchendo com a mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dia':["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    'Temperaturas':[30, np.nan, np.nan, 28, 27]
})

#Completa com o valor anterior as colunas vazias, sem dado
#bfill() - vai pegar o que estava embaixo
df_temperaturas['preenchido_bfill'] = df_temperaturas['Temperaturas'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome':["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    'cidade':["São Paulo", np.nan, "Curitiba", np.nan, "Belém"]
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna("Não informado")
df_cidades
```
